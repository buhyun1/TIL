### search_matched_sum / nlogn 상황


```python
def search_matched_sum(arr : list, target : int):
    low = 0 
    high = len(arr)-1
    min_index = []
    while (low < high) :
        sum = arr[low]+ arr[high]
        if sum == target:
            min_index = (low,high)
            break
        if target < sum:
            high -= 1
        else: 
            low += 1
    
    return min_index 
```

## 각 코드의 time complexity

### time complexity O(nlogn)
```python
def search_matched_sum(arr : list, target : int):
    low = 0  # 1
    high = len(arr)-1 # 1
    min_index = [] # 1 
    while (low < high) : # n 
        sum = arr[low]+ arr[high] # 1
        if sum == target:  
            min_index = (low,high) # 1
        if target < sum: #logn
            high -= 1 # 1
        else: #logn
            low += 1 # 1
    
    return min_index # 1
```

while 문에서 데이터 n 개를 돌며 그 코드 안에서 target과 sum을 if문으로 "<,>" binary(logn)으로 나눠줬으니 n * logn = O(nlogn)이다.
O(nlogn)

---

```python
def merge_sort(unsorted_array):
    if len(unsorted_array) > 1: # 1 
        mid = len(unsorted_array) // 2 # 1
        left = unsorted_array[:mid] # logn
        right = unsorted_array[mid:] # logn

        merge_sort(left) 
        merge_sort(right)

        i = j = k = 0
        
        while i < len(left) and j < len(right): # n
            if left[i] < right[j]: # 1
                unsorted_array[k] = left[i] # 1
                i += 1 # 1 
            else:
                unsorted_array[k] = right[j] # 1 
                j += 1 # 1
            k += 1 # 1

        while i < len(left): # n 
            unsorted_array[k] = left[i} # 1
            i += 1 # 1
            k += 1 # 1

        while j < len(right): # n
            unsorted_array[k] = right[j] # 1
            j += 1 # 1
            k += 1 # 1
    return unsorted_array
 ```
 
if 문 안에서 while문으로 데이터 n개를 돌고, 그 코드안에서 binary search로
left와 right를 slice를 통해 2개씩 나눠주기에 logn이기에 n*logn이다. 
o(nlogn)

### time complexity O(n^2)
```python
def search_matched_sum(arr : list, target : int):
  min_index = () # 1
  for i in range(0,len(arr)): # n
    for j in range(i+1,len(arr)): # n
      sum = arr[i] + arr[j] # 1
      if sum == target: # 1
        min_index = (i,j) # 1 
  return min_index      
```

for 문을 2번 돌며 i가 1일때 j가 (n-1)+ 2일때, j가 (n-2) + (n-3) ...+ 1 = 합이 n(n-1)/2 으로 나타나기에 빅오표기법에 의해 O(n^2)이다.
O(n^2)

---

```python
def ordered_search_matched_sum(arr : list): 
    
  for i in range(len(arr)): # n
      min_idx = i # 1
      for j in range(i+1, len(arr)): # n 
          if arr[min_idx] > arr[j]: # 1
              min_idx = j # 1

      arr[i], arr[min_idx] = arr[min_idx], arr[i] # 1
  return arr   
```

for 문을 2번 돌며 i가 1일때 j가 (n-1)+ 2일때, j가 (n-2) + (n-3) ...+ 1 = 합이 n(n-1)/2 으로 나타나기에 빅오표기법에 의해 O(n^2)이다.
O(n^2)

## 3.정렬에 따른 각 time complexity의 함수


### 각 time compleixty 에 해당하는 search_matched_sum 코드는 아래와 같다.

### 가설

정렬 하지 않고 n^2 합,


그다음은 정렬(nlogn) 후 nlogn 합, 


그다음은 정렬(nlogn) 후 n^2 합, 정렬(n^2) 후 nlogn 합이 동일 할 것이고,


마지막으로 정렬(n^2) 후 n^2 합 순서로 빠를 것이라 예상했다.


그 이유는


정렬 함수가 하나 없다면 그만큼 시간을 줄여줄 것이라 생각했고,


nlogn이 n^2보다 빠른 시간을 가지니 2,3등 순서는 따라 올것이고


(nlogn, n^2), (n^2 nlogn) 함수들은 n^2 정렬이 시간 복잡도를 보면 o(n^2)으로 동일 할 것으로 생각했기 때문이고,


마지막으로 n^2이 제일 오래 걸리니 이 순서대로 나올 것이라 예상했다.

# 성능 차이가 나는 이유

정렬(nlogn) 후 nlogn 합과 정렬(n^2) 후 n^2 합사이의 비교는 그래프를 통해 본다면 차이를 확실히 알 수 있기에


정렬(n^2) 후 nlogn 합, 정렬(nlogn) 후 n^2 합의 성능 차이가 나는 이유와


정렬(nlogn) 후 nlogn 합, 정렬 하지 않고 n^2 합의 가설과 실험결과가 다른 이유를 생각해보자면,


먼저, 정렬(n^2) 후 nlogn 합, 정렬(nlogn) 후 n^2 합의 성능 차이는 원하는 답을 검색을 통해 찾지 못해 
나의 실험 결과에 따른 분석을 해보자면 n이 커짐에 따라 정렬(nlogn) 후 n^2 합 함수가 점점 느려지고 있음을 보여줬다.  


n = 1000


n^2 sort = 8.571398000000045


nlogn sort = 1.0072714000125416


n^2 sum = 11.850866799999494


nlogn sum = 0.22116980000282638


### n^2 sort +  nlogn sort = 9.5786694


### n^2 sum  +  nlogn sum = 12.0720366


위의 표의 자료 와 비교 하면,


### 정렬(n^2) 후 nlogn 합 = 8.75420302


### 정렬(nlogn) 후 n^2 합 = 12.56101102	


오래 걸리는것은 sort냐 sum의 문제가 아닌 n^2, nlogn 의 차이였음에 성능차이가 났음을 알 수 있었다. 

---


그다음으로 정렬(nlogn) 후 nlogn 합, 정렬 하지 않고 n^2 합의 성능 차이는 


nlogn sort = 1.0072714000125416


nlogn sum = 0.22116980000282638


### nlogn sort +  nlogn sum = 1.2284412



### n^2 sum = 11.850866799999494

이것도 동일하게 n^2와 nlogn의 차이였음에 성능차이가 났음을 알 수 있다.
