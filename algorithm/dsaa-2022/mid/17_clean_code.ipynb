```python
class ClassWithRegularAttributes:
    def __init__(self, someParameter):
        self.someAttribute = someParameter
        
# a.someAttribute = b
```

```python
obj = ClassWithRegularAttributes('some initial value')
print(obj.someAttribute) # some initial value 출력
obj.someAttribute = 'changed value'
print(obj.someAttribute) # changed value 출력
del obj.someAttribute # someAtrribute 속성 삭제
```

```python
class ClassWithProperties:
    def __init__(self):
        return self._someAttribute #'some initial value'
    
    @property
    def someAttribute(self): # getter 메소드
        return self._someAttribute
    
    @someAttribute.setter
    def someAttribute(self, value): # setter 메소드
        self._someAttribute = value
    
    @someAttribute.deleter
    def someAttribute(self): # deleter 메소드
        del self._someAttribute
```

```python
obj = ClassWithRegularAttributes('some initial value')
print(obj.someAttribute)
obj.someAttribute = 'changed value'
print(obj.someAttribute) 
del obj.someAttribute 
```

```python
#속성을 프로퍼티로 바꾸기
class ClassWithProperties:
    def __init__(self):
        self.someAttribute = 'some initial value'
    
    @property
    def someAttribute(self): # getter 메소드
        return self.someAttribute
    
    @someAttribute.setter
    def someAttribute(self, value): # setter 메소드
        self.someAttribute = value
        
obj = ClassWithProperties()
print(obj.someAttribute) #dead kernel #information hiding
```

```python
#데이터 검증 위한 setter 사용
class WizCoinException(Exception):
    "예외 발생"
    pass # 상속으로 error 발생

class WizCoin:  
    def __init__(self, galleons, sickles, knuts):  
        """galleons, sickles, knuts로 새로운 WizCoin 객체를 생성한다."""
        self.galleons = galleons 
        self.sickles = sickles
        self.knuts = knuts # init 메소드는 반환문이 없다.
        
    @property
    def galleons(self):
        return self._gallons
    
    @galleons.setter
    def galleons(self, value):
        if not isinstance(value, int): # int 인지 확인
            raise WizCoinException('gallons attr must be set to an int, not a' + value.__class__.__qualname__)
        if value < 0: # 양수인지 확인
            raise WizCoinException('galleons attr must be a positive int, not' + value.__class__.__qualname__) #객체 이름의 문자열 획득
            self._gallons = value 
                   
```

```python
#읽기전용 프로퍼티
@property
def total(self):
    return (self.galleons * 17 * 29) + (self.sickles * 29) + (self.knuts) # total = a 로 수정불가 # galleon, sickle, knut에 의존

#total = a -> total = b ,포인터
#galleon, sickle, knut 변경
```
