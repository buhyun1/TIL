```python
import csv
from collections import defaultdict

f = open("forklist_move.csv","r") #     
rdr = csv.reader(f)  

list_dict = defaultdict(list) #defaultdict list .
for i in rdr:
  list_dict[i[1]].append([i[3],i[4],i[2]])
del(list_dict['fork_id']) #   dict  
#    

list_dict1 = sorted(list_dict['TEAM10054239'], key = lambda x: x[2])
list_dict2 = sorted(list_dict['TEAM1007B9625'], key = lambda x: x[2]) 
list_dict3 = sorted(list_dict['TEAM10083967'], key = lambda x: x[2]) 

list_dict['TEAM10054239'] += list_dict1
list_dict['TEAM1007B9625'] += list_dict2
list_dict['TEAM10083967'] += list_dict3

f.close()
print(list_dict)
```
