```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/auto.csv')
```

```python
df
```

```python
#전처리
#origin 변수 onehot encoding
df = pd.get_dummies(df, columns = ['origin'])
df
```

```python
#독립, 종속변수 나누기
y = df['mpg']
x = df.drop(columns = ['mpg'])
```

```python
# 회귀분석
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
#학습된 회귀모형 평가
 # 학습 사용된 데이터 -> 학습된 모형 -> 예측값 = y_pred
 # 학습 사용된 데이터 정답 = y
 # rmse 
 # r2-score
```

```python
#rmse
y_pred = reg.predict(x)
y_pred #numpy 배열
```

```python
y #pandas series
```

```python
#rmse
import numpy as np
rmse = np.sqrt(np.sum(((y_pred - y)**2)/y.count()))
rmse
```

```python
#r2score
y_mean = y.mean()
y_mean
```

```python
tss = np.sum((y- y_mean)**2)
tss
```

```python
rss = np.sum((y-y_pred)**2)
rss
```

```python
r2score = 1- rss/tss
r2score
```

```python
#sklearn 패키지 이용 평가지표 구하기
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.sqrt(mean_squared_error(y,y_pred)) #rmse
```

```python
r2_score(y,y_pred)
```

```python
#mae
np.mean(np.abs(y- y_pred))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y,y_pred)
```

```python
#mape
np.mean(np.abs((y-y_pred)/y))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y,y_pred)
```

```python
z=y*0.01 # 정답 데이터의 스케일을 100분에 1로 낮춤
z_pred = y_pred*0.01 # 예측값의 스케일을 100분에 1로 낮춤
```

```python

print('mse: ',mean_squared_error(z,z_pred))
print('r2_score: ',r2_score(z,z_pred)) #단위 영향x
print('mae: ',mean_absolute_error(z,z_pred))
print('mape: ',mean_absolute_percentage_error(z,z_pred)) #단위 영향x
```

```python
#mape 값이 이상하게 나오는경우
z[0] = 0.0000000000000000001 # 실제 레이블의 정답이 0에 가까운 경우 주의해야함z
```

```python

print(mean_squared_error(z,z_pred))
print(r2_score(z,z_pred))
print(mean_absolute_error(z,z_pred))
print(mean_absolute_percentage_error(z,z_pred)) #mape 큰 값나옴
#식보면 이해 가능
```
