```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/heart.csv')
df
```

```python
categorical_features = ['Sex','ChestPain','Thal','AHD'] #원핫인코딩할 변수선정
```

```python
df2 = pd.get_dummies(df, columns = categorical_features) # 원핫인코딩 수행
```

```python
df3 = df2.dropna(axis=0)
df3 # 누락치 제거거
# 범주형 데이터는 원핫인코딩 하고 머신러닝 시작해야함
```
