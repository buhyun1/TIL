```python
#로지스틱함수
import numpy as np
```

```python
z = np.linspace(-10,10,100)
z
```

```python
## 로지스틱 함수 정의
def sigmoid(x):
  return 1/(1+np.exp(-1*x))
```

```python
## 함수 작동 확인
## numpy array의 broadcasting , 벡터의 값들을 함수에서 처리해서 다시 벡터로, 차원 확장하는 개념
sigmoid(z)
```

```python
plt.plot(z,sigmoid(z))
```

```python
# 로지스틱회귀 학습
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
#데이터 분할(독립,종속)
```

```python
x=df[['X1','X2']]
y=df['y']
```

```python
reg.fit(x,y)
```

```python
#학습결과 확인(패러미터)
```

```python
W = reg.coef_ #기울기
W 
```

```python
b = reg.intercept_ #절편
b
```

```python
## 학습된 패러미터로 계산한 확률 = sklearn에서 계산한 확률
```

```python
reg.predict_proba(x)[:,:]
```

```python
#학습된 모델에 학습에 사용된 데이터 각각을 넣어서 각각이 0또는 1로 분류될 확률 계산

reg.predict_proba(x)[:,0] # 0이 될 확률 / 1번째 열 정보 가져와라
reg.predict_proba(x)[:,1] # 1이 될 확률 / 모든 행에 대해 2번째 열 정보 가져다 달라
# 이진분류 문제일 때 [0, 1]사이에서 곧 1일 확률(0이 아닐 확률)
# 첫 번째 원소 = 첫 번째 클래스의 예측 확률
# 두 번째 원소 = 두 번째 클래스의 예측 확률
# 즉 두 클래스에 대한 확률의 합( 첫 번째 클래스의 예측 확률 + 두 번째 클래스의 예측 확률)은 항상 1이다.
```

```python
sigmoid(np.dot(x,W.T)+b)  #뒤집어야 (100,1)나옴
# 선형회귀로 예측되는 값 

# numpy.dot은 행렬의 곱을 표현
# 위에 결과와 동일한 값
```
