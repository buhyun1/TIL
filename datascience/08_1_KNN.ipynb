```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tonyfischetti/InteractiveLogisticRegression/master/moons.csv')
df
```

```python

plt.scatter(df['X1'],df['X2'],c=df['Y'])
```

```python
X=np.array(df[['X1','X2']])
y=np.array(df['Y'])
X.shape, y.shape
```

```python
x = np.array([1,-1]) # 새로운 데이터셋 나중에 쓸것 (X1,X2) = (1,-1)
x
```

```python
a=np.array([3,3])
b=np.array([-2,-1.5])
np.sqrt(np.sum((a-b)**2))
```

```python
result = np.sqrt(np.sum((x-X)**2,axis=1)) # (1,-1)과 X사이의 거리가 순서대로 적혀있음 
#(1,-1) 과 (0.997072, -0.764072) 사이 거리 계산
result
```

```python
# argsort 예시
# 리스트안에 숫자를 정렬한다고 했을때 인덱스 기준으로 어떻게 나와야 되냐
k = np.array([1,3,2,100,20])
k[np.argsort(k)]
```

```python
# [-1,1]과 모든 데이터 사이의 거리의 argsort -> order
order = result.argsort()
order
```

```python
X[order][:5] 
```

```python
# 가까운 순위가 5보다 작은것들
X[order[:5]] # order 정렬 되있는데 , 5개까지 가져와서 x 값 가져오기
```

```python
y[order[:5]].mean()
```

```python
y[order[:100]].mean()
```

```python
from sklearn.metrics import pairwise_distances
# 모든 데이터 사이의 거리
dist = pairwise_distances(x.reshape(1,-1),X)[0]
dist
```

```python
y[dist.argsort()[:5]].mean()
```
