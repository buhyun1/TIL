```python
import numpy as np
```

```python
true_w = 2
true_b = -3
```

```python
x = np.random.normal(1, size = 50) #평균 0, 표편 1 정규분포로 부터 50개 데이터 무작위 추출
x
```

```python
y = true_w*x +true_b 
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
noise = np.random.normal(0.1,size=50)
noise
```

```python
y = y + noise
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y) # error 당연히 나옴
```

```python
x=x.reshape(-1,1) 
#linear regression의 fit 함수의 x 인자는 2차원 벡터 요구
# 현재 x는 1차원 reshape 이용하여 2차원으로 바꾸기
# -1의 의미 남은 배열의 길이와 남은 차원으로부터 추정해서 알아서 행을 지정
x
# 아래 형식이 2차원 벡터 형식 [ [],[],[] ]
```

```python
reg.fit(x,y)
```

```python
reg.coef_ 
# 기울기: reg.coef_
# 절편: reg.intercept_
```

```python
reg.intercept_
```

```python
reg.predict(x) #y의 예측값
```

```python
y_pred = reg.predict(x)
y_pred 
```

```python
plt.scatter(x,y) #원래 데이터
plt.scatter(x,y_pred) #예측된 데이터
```

```python
# 새로운 데이터
x2= np.linspace(-3,3,50) # -3~3까지 50개로 나눔
y2_pred = reg.predict(x2.reshape(-1,1))
y2_pred
```
