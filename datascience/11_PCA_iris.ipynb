```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris(as_frame=True)
```

```python
iris['target_names']
```

```python
df=iris['data']
df
```

```python
species=iris['target']
species
```

```python
df_norm = (df-df.mean())/df.std()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=4)
Z=pca.fit_transform(df_norm) # fit 과 transform 이 합쳐져서 한번에 진행 학습과 예측이 동시에
Z.shape
```

```python

pca.explained_variance_
```

```python

pca.explained_variance_ratio_
```

```python
import numpy as np
np.cumsum(pca.explained_variance_ratio_) #누적
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.scatter(Z[:,0],Z[:,1],c=species) #첫번째, 두번쨰 열
```

```python
factor=pca.components_
factor # 변환된 차원에서 0.52, 0.37 씩 증가한다.
# 2개로 압축 됐으니 2행까지만 사용
```

```python
factor=pca.components_[:2,:].T # 고유행렬에서 첫 2행까지의 데이터 추려내기
plt.figure(figsize=(12,8))
plt.scatter(Z[:,0],Z[:,1],c=species)
for i in range(4):
    plt.annotate(df.columns[i],xy=(0,0),xytext=(factor[i]*2),arrowprops={'lw':1.5, 'arrowstyle':'<-'})
```
