```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris(as_frame=True)
```

```python
df=iris['data']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_norm = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3) #kmeans 클러스터링 객체 생성
kmeans.fit(df_norm) # 학습
```

```python

kmeans.cluster_centers_ #그룹 별 중점 나옴
```

```python
## 중점과 소속된 점들사이 평균거리 확인
kmeans.inertia_
```

```python
# 분류된 점 확인
kmeans.labels_
```

```python
kmeans.predict([[0,0,0,0]])
```

```python
cn=[]
measure=[]
for i in range(1,10):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(df)
    cn.append(i)
    measure.append(kmeans.inertia_)
```

```python

measure
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(cn,measure,'ro-')
plt.bar(cn,measure,tick_label=cn)
```
