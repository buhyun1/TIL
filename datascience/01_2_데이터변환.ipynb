```python
df =pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/classification_data.csv')
df
```

```python
x = df[['X1','X2']]
x
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(3, include_bias = False) #3차식
# degree 옵션으로 차수를 조절한다.
# include_bias 옵션은 True로 할 경우 0차항(1)도 함께 만든다.
# 입력 샘플이 2차원이고 [a, b] 형식인 경우 2차 다항식 기능은 [1, a, b, a^2, ab, b^2]

# PolynomialFeatures 클래스는 입력 데이터 x를 [1,x,x2,x3,...]과 같이 여러 다항식으로 변환한다.
output = poly.fit_transform(x)
output
```

```python
c= poly.get_feature_names_out(x.columns)
# get_feature_names_out() 메서드를 호출하면 각 특성이 각각 어떤 입력의 조합으로 만들어졌는지 알려준다. 

# 예시
# print(poly.get_feature_names_out()) 
# ['x0' 'x1' 'x2' 'x0^2' 'x0 x1' 'x0 x2' 'x1^2' 'x1 x2' 'x2^2']
pd.DataFrame(output, columns=c)
```
